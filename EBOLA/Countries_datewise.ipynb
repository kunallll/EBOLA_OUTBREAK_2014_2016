```python
import pandas as pd 

df = pd.read_csv('ebola_2014_2016_clean.csv')
```

```python
filt = df['Country'] == 'Guinea'
import matplotlib.pyplot as plt1
import numpy as np

x_indexes = np.arange(df[filt]['Date'].count())
plt1.title('GUINEA')
plt1.xlabel('Days')
plt1.ylabel('Suspected Cases')
plt1.plot(x_indexes, df[filt]['No. of suspected cases'], label = 'Suspected Cases')
plt1.plot(x_indexes, df[filt]['No. of probable cases'], label = 'Probable Cases')
plt1.plot(x_indexes, df[filt]['No. of confirmed cases'], label = 'Confirmed Cases')
plt1.legend()
```
